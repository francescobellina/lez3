#confronto tra diverse configurazioni di binning : aumentare o diminuire il numero dei bin
## https://en.wikipedia.org/wiki/Data_binning

```python
#!/usr/bin/python
'''
example on the use of matplotlib.pyplot
with the comparison among different binning choices
'''

import matplotlib.pyplot as plt
import numpy as np
from math import ceil, floor #floor invece è proprio la funzione "parte intera"


def sturges (N_events) :
    return ceil (1 + 3.322 * np.log (N_events)) #regola di sturges


# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----


def main () :
    '''
    Funzione che implementa il programma principale
    '''

    # the sample to be represented as an histogram
    with open('dati.txt') as f:
        sample = [float (x) for x in f.readlines()]

    xMin = floor (min (sample)) #prende la parte intera del valore minimimo (la floor approx sempre all'intero inf)
    xMax = ceil (max (sample)) #prende la parte intera superiore del massimo dei valori di input
    N_bins = sturges (len (sample)) #len(sample) equivale al numero di eventi
    fig, axes = plt.subplots (nrows = 3, ncols = 1) #ho 3 grafici, incolonnati uno sull'altro

    bin_edges = np.linspace (xMin, xMax, N_bins) #operazione di binning dell'intervallo x
    axes[0].hist (sample,  #PRIMO GRAFICO ( axes[0])
             bins = bin_edges,
             color = 'orange',
             label = 'Sturges',
            )

    bin_edges = np.linspace (xMin, xMax, N_bins * 10) #in questo caso divido l'intervallo di 10 volte di piu (ho bin piu stretti e tanti)
    axes[1].hist (sample,    #SECONDO GRAFICO : qui avrò un'approx migliore
             bins = bin_edges,
             color = 'blue',
             label = 'Sturges x 10',
            )


    bin_edges = np.linspace (xMin, xMax, int (N_bins / 10)) #qui faccio l'esatto opposto
    axes[2].hist (sample,  #TERZO GRAFICO
             bins = bin_edges,
             color = 'red',
             label = 'Sturges / 10',
            )


    for ax in axes : #cosi non devo farlo per tutti e tre i grafici
        ax.set_xlabel ('variable')
        ax.set_ylabel ('event counts')
        ax.legend ()

    plt.savefig ('drawing_06.png')
    plt.show ()


# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----


if __name__ == "__main__":
    main ()
```

#Scale logaritmiche
##FUNZIONE GAUSSIANA

```python
#!/usr/bin/python
'''
example on the use of matplotlib.pyplot
and the use of the logarithmic scale on the vertical axis
'''

import matplotlib.pyplot as plt
import numpy as np
from math import sqrt


def Gaussian (x, mean, sigma) :   #FUNZIONE GAUSSIANA
    if sigma == 0 : return float (x == 0.)
    return np.exp (-0.5 * pow ((x - mean) / sigma, 2.)) / (sqrt (2 * np.pi) * sigma) #forma analitica


# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----


def main () :
    '''
    Funzione che implementa il programma principale
    '''

    # preparing the set of points to be drawn
    x_coord = np.linspace (-3., 3., 10000)
    y_coord = np.arange (0., x_coord.size)

    for i in range (x_coord.size):
        y_coord[i] = Gaussian (x_coord[i], 0., 1.) #per ciascun punto ...

    fig, axes = plt.subplots (nrows = 1, ncols = 2) #due grafici di fianco

    axes[0].plot (x_coord, y_coord)
    axes[0].set_title ('linear y scale', size=14)
    axes[0].set_xlabel ('x')
    axes[0].set_ylabel ('y')

    axes[1].plot (x_coord, y_coord)
    axes[1].set_title ('log y scale', size=14)
    axes[1].set_xlabel ('x')
    axes[1].set_yscale ('log')

    plt.savefig ('drawing_07.png')
    plt.show ()


# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----


if __name__ == "__main__":
    main ()
```

#CALCOLO DEI MOMENTI DA UN CAMPIONE
##media, varianza ecc andrebbero salvate su una apposita libreria stats

```python
##di seguito è la libreria stats
from math import sqrt

def mean (sample) : #SAMPLE è IL CAMPIONE DEI DATI
    '''
    calculates the mean of the sample present in the object
    '''
    summ = sum (sample)
    N = len (sample)
    return summ / N

# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----

def variance (sample, bessel = True) :
    '''
    calculates the variance of the sample present in the object
    '''
    summ = 0.
    sum_sq = 0.
    N = len (sample)
    for elem in sample :
       summ += elem
       sum_sq += elem * elem
    var = sum_sq / N - summ * summ / (N * N)
    if bessel : var = N * var / (N - 1)
    return var

# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----

def sigma (sample, bessel = True) :
    '''
    calculates the sigma of the sample present in the object
    '''
    return sqrt (variance (sample, bessel))

# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----

def sigma_mean (sample, bessel = True) :
    '''
    calculates the sigma of the sample present in the object
    '''
    N = len (sample)
    return sqrt (variance (sample, bessel) / N)
#-------------------------
#PROGRAMMA PRINCIPALE (PER IL CALCOLO DEI MOMENTI):
#!/usr/bin/python
'''
example on saving information on txt files
'''

import sys #HO modificato l'esempio perchè usava sys per la lettura dei dati (guardare gli esempi della lezione3)
import numpy as np

def main () :
    '''
    Funzione che implementa il programma principale
    '''
    # read the file
    sample=np.loadtxt('dati.txt')

    print (len (sample))

    print ('sample mean:', mean (sample)) ;
    print ('sample variance:', variance (sample)) ;


# ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ---- ----


if __name__ == "__main__":
    main ()
```
